```python
import requests
from bs4 import BeautifulSoup
```

```python
page_count=1

while True:
    URL=f"https://quotes.toscrape.com/page/{page_count}/"
    res=requests.get(URL)

    Soup=BeautifulSoup(res.text, "lxml")
    quotes=Soup.select("div.quote")

    if not quotes:
        print("not valid pages anymore...")
        break

    with open(f"scraped_data/pages{page_count}.html", "w", encoding="UTF-8") as f:
        f.write(res.text)
        print(f"pages downloaded successfully{page_count}")

    page_count=page_count+1
        
```

```python
import os
import time

page_count=1
life_quotes=[]

headers = {
    "User-Agent": "Mozilla/5.0"
}

while True:
    URL=f"https://quotes.toscrape.com/page/{page_count}/"

    try:
        res=requests.get(URL)
    except requests.exceptions.RequestException as e:
        print("Network error:", e)
        break

        
    soup=BeautifulSoup(res.text, "lxml")
    all_quotes=soup.select("div.quote")

    if not all_quotes:
        break

    for q in all_quotes:
            all_tags=[tag.get_text(strip=True) for tag in q.select(".tag")]
    
            if "life" in all_tags:
             text=q.select_one("span.text").get_text()
             author=q.select_one("small.author").get_text()
             life_quotes.append([text, author])
    page_count=page_count+1
    time.sleep(1)
    
print(life_quotes)
```
