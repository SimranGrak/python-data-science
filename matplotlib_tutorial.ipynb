```python
# Movies Data

# Oscar Winner Movies
oscar_movies = [
    "The Dark Knight", 
    "The Hurt Locker",  
    "The King's Speech", 
    "The Artist",     
    "Argo"
]
oscar_years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

plt.plot(oscar_years, oscar_revenue)
plt.title("Oscar movie revenue in each year")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
plt.show()
```

```python
# Non-Oscar Winner Movies
non_oscar_movies = ["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_years = [2008, 2009, 2010, 2011, 2012]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M

plt.plot(non_oscar_years, non_oscar_revenue,"o--r", label="Non-oscar winner")
plt.plot(oscar_years, oscar_revenue,"<-b", label="Oscar winner")
plt.title("Years vs Revenue(in $M)")
plt.xlabel("Years")
plt.ylabel("Revenue in ($M)")

plt.legend()
```

```python
# Scatter Plots
people = ["Person A", "Person B", "Person C", "Person D", "Person E", 
          "Person F", "Person G", "Person H", "Person I", "Person J"]
age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]
bp = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]

colors=["green" if x<135 else "red" for x in bp]
plt.scatter(age, bp, cmap="OrRd",c=bp, s=bp, marker="o", alpha=0.9)
plt.title("age vs bp")
plt.xlabel("age in years")
plt.ylabel("bp")

plt.grid()
plt.colorbar()


for i in range(len(people)):
    plt.annotate(people[i], xy=(age[i], bp[i]), xytext=(age[i]+2, bp[i]+0.5))


plt.xlim(min(age), max(age)+20)
plt.ylim(min(bp), max(bp)+5)
```

```python
# Multiple scatter plots
cities = ["City A", "City B", "City C", "City D", "City E"]

# Winter: Temperature (C) vs Humidity (%)
winter_temp = [5, 2, 10, 0, 7]
winter_humidity = [80, 75, 65, 85, 70]

# Summer: Temperature (C) vs Humidity (%)
summer_temp = [25, 30, 28, 35, 27]
summer_humidity = [60, 50, 55, 45, 65]

plt.scatter(winter_temp, winter_humidity, label="winter")
plt.scatter(summer_temp, summer_humidity, label="summer")
plt.xlabel("temperature()")
plt.ylabel("humidity(%)")
plt.title("cities temperature")
plt.legend()
```

```python
# Pie Charts
expenses = ["Salaries", "Rent", "Marketing", "R&D", "Miscellaneous"]
amounts = [500, 150, 120, 100, 50]

explode=[0,0,0,0.2,0]

plt.style.use("classic")
plt.pie(amounts, labels=expenses, autopct="%1.1f%%",
       wedgeprops={
           "edgecolor":"black",
           "linewidth":1,
           "linestyle":"--"
       }, 
       explode=explode,
        startangle=360,
        shadow=True
       )
plt.show()
# plt.style.available
```
