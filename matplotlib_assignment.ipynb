```python
#matplotlib assignment
import matplotlib.pyplot as plt
import numpy as np

'''For the following dataset generated using NumPy:
np.random.normal(loc=70, scale=10, size=200)
a) Plot a histogram of scores 
b) Draw a vertical line for the mean & try to interpret the results'''

scores=np.random.normal(loc=70, scale=10, size=200)

plt.hist(scores ,color="pink", bins=15, edgecolor="black")
plt.xlabel("marks")
plt.ylabel("no. of students")
plt.title("marks of students")

mean_scores=np.mean(scores)

plt.axvline(x=mean_scores,
           color="red",
    linestyle="--",
    linewidth="2",
    label="mean score")

plt.legend()
plt.show()
```

```python
'''For the Seaborn penguins dataset ,do the following:
a) Plot flipper_length_mm vs body_mass_g
b) Color by species'''

import seaborn as sns

sns.get_dataset_names()

penguins=sns.load_dataset("penguins")

sns.relplot(
    data=penguins,
    x="flipper_length_mm",
    y="body_mass_g",
    hue="species"
)
```

```python
'''For the Seaborn tips dataset, do the following:
a) create a scatter plot of total_bill vs tip
b) color points by sex
c) use different marker for smoker
d) add a title and axis labels
e) analyze: Do smokers tend to tip differently
'''

tips=sns.load_dataset("tips")

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="sex",
    style="smoker"
)
plt.title("total_bill vs tips")
plt.xlabel("total_bill")
plt.ylabel("tips")

print(tips.head())

#Conclusion: Smoking status does not appear to have a strong impact on tipping behavior.
```

```python
'''
for the seaborn tips dataset, do the following:
a) create a pivot table of average tip
    1. Rows: day
    2. columns: time
c) plot the pivot table as a heatmap
d) annonate values
e) Interpret which day-time combination performs best
'''

tips=sns.load_dataset("tips")

pivot_table=tips.pivot_table(values="tip", index="day", columns="time", aggfunc="mean")

# Plot heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(
    pivot_table,
    annot=True,      # annotate values
    fmt=".2f",       # show 2 decimal places
    cmap="YlOrRd"    # color map
)

# Add labels and title
plt.title("Average Tip by Day and Time")
plt.xlabel("Time")
plt.ylabel("Day")

# Show plot
plt.show()

print(tips.head())
```
